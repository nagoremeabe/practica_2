```python
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
import cartopy.io.shapereader as shpreader
```

```python
# Downloaded from http://biogeo.ucdavis.edu/data/gadm2/shp/DEU_adm.zip
fname = './mapas/gadm41_ESP_1.shp'

adm1_shapes = list(shpreader.Reader(fname).geometries())
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(14, 10), ncols=1)
ax=plt.subplot(1,1,1,projection=ccrs.Robinson())

plt.title('Spain')
ax.coastlines(resolution='10m')

ax.add_geometries(adm1_shapes, ccrs.PlateCarree(), # Cuidado con los sistemas de coordenadas, esto esta definido solo para este sistema
                  edgecolor='black', facecolor='w', alpha=0.5)


ax.set_global() # added following an answer to my question
#ax.gridlines()
ax.gridlines(draw_labels = True,linestyle='--',alpha=0.7,zorder=1)
def pinta_region(lats,lons,color,ax=ax,label=None):
    lat_corners = np.array([lats[0], lats[0],lats[1], lats[1]])
    lon_corners = np.array([ lons[0], lons[1], lons[1], lons[0]])
    poly_corners = np.zeros((len(lat_corners), 2), np.float64)
    poly_corners[:,0] = lon_corners
    poly_corners[:,1] = lat_corners
    poly = mpatches.Polygon(poly_corners, closed=True, ec=color, fill=True, lw=1, fc=color,label=label,alpha=0.6, transform=ccrs.Geodetic()) # geodetic para que transforme los rectngulos a la proyeccin correcta
    ax.add_patch(poly)

ax.set_title('Regiones de inters',size=22)

pinta_region([42,43.8],[-9,-6],'royalblue',ax,label='Oeste')
pinta_region([42.5,43.8],[-6,0],'seagreen',ax,label='Norte')
pinta_region([38, 42],[-7, -1],'goldenrod',ax,label='Centro')
pinta_region([36, 38],[-8, 0],'orangered',ax,label='Sur')
pinta_region([36.2, 43],[-2, 4],'purple',ax,label='Este')
#pinta_region([0,60],[-70,0],'seagreen',ax,label='Atlntico')
#pinta_region([-30,35],[40,110],'brown',ax,label='ndico')
ax.legend(prop={'size':12})




ax.set_extent([-9.8,5,35.5,44], ccrs.PlateCarree())
fig.savefig('mapa_regiones_ESP.png',dpi=600,facecolor='w')
```

```python
%matplotlib qt
fig, ax = plt.subplots(figsize=(14, 10), ncols=1)
ax=plt.subplot(1,1,1,projection=ccrs.Robinson())

plt.title('Spain')
ax.coastlines(resolution='10m')

ax.add_geometries(adm1_shapes, ccrs.PlateCarree(), # Cuidado con los sistemas de coordenadas, esto esta definido solo para este sistema
                  edgecolor='black', facecolor='w', alpha=0.5)


ax.set_global() # added following an answer to my question
#ax.gridlines()
ax.gridlines(draw_labels = True,linestyle='--',alpha=0.7,zorder=1)
def pinta_region(lats,lons,color,ax=ax,label=None):
    lat_corners = np.array([lats[0], lats[0],lats[1], lats[1]])
    lon_corners = np.array([ lons[0], lons[1], lons[1], lons[0]])
    poly_corners = np.zeros((len(lat_corners), 2), np.float64)
    poly_corners[:,0] = lon_corners
    poly_corners[:,1] = lat_corners
    poly = mpatches.Polygon(poly_corners, closed=True, ec=color, fill=True, lw=1, fc=color,label=label,alpha=0.6, transform=ccrs.Geodetic()) # geodetic para que transforme los rectngulos a la proyeccin correcta
    ax.add_patch(poly)

ax.set_title('Regiones de inters',size=22)

#pinta_region([42,43.8],[-9,-6],'royalblue',ax,label='Oeste')
pinta_region([41.8,43.8],[-9.3,-1],'seagreen',ax,label='Norte')
pinta_region([38.8, 41.8],[-7.5, -1],'goldenrod',ax,label='Centro')
pinta_region([36, 38.8],[-8, -1],'orangered',ax,label='Sur')
pinta_region([37.5, 43],[-1, 4.3],'purple',ax,label='Este')
#pinta_region([0,60],[-70,0],'seagreen',ax,label='Atlntico')
#pinta_region([-30,35],[40,110],'brown',ax,label='ndico')


ax2 = fig.add_axes([0.8, 0.8, 0.2, 0.2], projection=ccrs.PlateCarree())
ax2.set_global()
ax2.coastlines(resolution='10m')
ax2.add_geometries(adm1_shapes, ccrs.PlateCarree(), # Cuidado con los sistemas de coordenadas, esto esta definido solo para este sistema
                  edgecolor='black', facecolor='aquamarine', alpha=0.5)


ax2.set_global() # added following an answer to my question
#ax.gridlines()
ax2.gridlines(draw_labels = False,linestyle='--',alpha=0.5,zorder=1)
ax2.set_extent([-13,-18.5,27,30], ccrs.PlateCarree())

def reposition():
    plt.draw()
    p1 = ax.get_position()
    p2 = ax2.get_position()
    ax2.set_position([p1.x1-1.1*p2.width, p1.y0-0.5*p2.height, p2.width, p2.height])

reposition()
pinta_region([34.5,44],[-10,7],'aquamarine',ax=ax2,label='Canarias')


ax.set_extent([-10,7,34.5,44], ccrs.PlateCarree())
lgd = ax.legend(prop={'size':12},loc=1)

def add_patch(legend):
    from matplotlib.patches import Patch
    ax = legend.axes

    handles, labels = ax.get_legend_handles_labels()
    handles.append(Patch(facecolor='aquamarine'))
    labels.append("Canarias")

    legend._legend_box = None
    legend._init_legend_box(handles, labels)
    legend._set_loc(legend._loc)
    legend.set_title(legend.get_title().get_text())


add_patch(lgd)
plt.show()

#fig.savefig('mapa_regiones_ESP.png',dpi=600,facecolor='w')

```
